## **Basic rules of Probability**

- complement rule
- addition rule
- multiplication rule

#### **complement rule** : The event not occuring

```python
# P(A) = 1 - P(B)

# eg : P(no rain) = 1-P(rain)
# if p(rain) = 30%, then p(no rain) = 1-p(0.3) = 0.7 or 70%
```

> Mutually exclusive events: 
- Two events are mutually exclusive if they cannot occur at the same time

#### **Addition Rule**(for mutually exclusive events):


```python
# P(A OR B) = P(A) + P(B)

# eg : P(rolling a 2 or rolling a 4) = P(rolling a 2) + P(rolling a 4)
# P(rolling a 2 or rolling a 4) = 1/6 + 1/6 = 2/6 = 1/3 or 33% 
```

> Independent events:
- Two events are independent if the occurrence of one event does not change the probability of the other event

#### **Multiplication Rule** (for indipendent variables)

```python
# P(A AND B) = P(A) * P(B)

# eg : P(tail and head) = P(tails) * P(heads)
# P(tails and heads) = (0.5) * (0.5) = 0.25 = 25%
```

## **Conditional Probability**

The probability of an event occuring given that another event has already occurred 

> Dependent events: Two events are dependent if the occurance of one event changes the probility of the other event 

```python
# P(A AND B) = P(A) * P(B|A)   9R P(B|A) = P(A AND B) / P(A)

# P(A) = ACE ON FIRST DRAW = 4/52
# P(B|A) = ACE ON SECODN DRAW = 3/51

# P(A AND B) = 4/52 * 3/53 = 1/221 = 0.05%
```

```python
# A = GETTING ACCEPTED (10% 10 OUT OF 100)
# B = GETTING A SCHOLARSHIP (2% 2 OUT OF 100)
# P(A AND B) = PROB OF GETTING ACCEPTED AND GETTING SCHOLARSHIP

# P(A AND B) = 10/100 * 2/100 = 1/500 = 0.2%
```

## **Bayes Theorem**

> P(A|B) = P(B|A) * P(A) / P(B)

* P(A) = Prior Probability: The probability of an event before new data is collected
* p(A|B)= Posterior Probability: The updated probability of an event based on new data

#### **Probability Calculation** ( Eg : Weather Information)
* overall chance of rain = 10%
* all days starts of cloudy = 40%
* rainy days starts of cloudy = 50%
#### **Probabilities**
* Prior probability = the probability of a rainy day
* Posterior probability = the probability that it will rain given that it's cloudy

```python
# event A = Rain
# evnet B = Cloudy
# P(RAIN | CLOUDY) =  P(CLOUDY|RAIN) * P(RAIN) / P(CLOUDY)

# P(RAIN) = 10%
# P(CLOUDY) = 40%
# P(CLOUDY|RAIN) = 50%

# P(RAIN | CLOUDY) = (0.5 * 0.1) / 0.4 = 0.125 = 12.5%
```

> IN CERTAIN SITUATIONS LIKE FALSE POSITIVE AND FALSE NEGATIVE WE USE THE EXPANDED VERSION OF BAYES THEOREM :

```python
# P(A) = 1% (PERSON HAS THE ALLERGY)
# P(B|A) 95% (+VE TEST GIVEN THAT THE PERSON HAS THE ALLERGY)
# P(B|NOT A) = 2% (+VE TEST GIVEN THAT THE PERSON DOES NOT HAVE THE ALLERGY)
# P(NOT A) = 1-P(A) ie prob of event A does not occur = 1-0.01 = 0.99 =99%

# P(A|B) = 0.95 * 0.01 / 0.95 * 0.01 + 0.02 * 0.99 = 0.324 = 32.4%

# P(A | B) IS 32% THAT MEANS ITS NOT VERY LIKELY THAT A RANDOM PERSON WILL BOTH TEST +VE AND HAVE ALLERGY
```

#### **Binomial Distribution** : 
> A discrete distribution that models the probability of events with only two possible outcomes, success or failure

Binomial experiment: 
* 10 repeated coin tosses
* Two possible outcomes: heads or tails
* The prob of success for each toss is the same : 50%

**Formula :**
P(X=k) = nCk * pk * (1-p)n-k

**Implementation of binomial distribution Using Python :**

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
n = 5  # number of draws
k = 0  # number of successes we are interested in
p = 13 / 52  # probability of drawing a heart (13 hearts in a 52-card deck)

# Compute the binomial probability
probability = binom.pmf(k, n, p)

# Print the result
print(f"Probability of drawing exactly {k} hearts in {n} draws: {probability:.4f}")

# To visualize the binomial distribution for different numbers of hearts
k_values = np.arange(0, n + 1)
probabilities = binom.pmf(k_values, n, p)

plt.bar(k_values, probabilities, tick_label=k_values,color='b')
plt.xlabel('Number of Hearts')
plt.ylabel('Probability')
plt.title('Binomial Distribution of Drawing Hearts')
plt.grid(True)
plt.show()
```

#### **Poisson Distribution** 

> Models the probability that a certain number of events will occur during a specific time period or distance

Poisson experiment:
* The number of evnets in the experiment can be counted 
* the mean number of events that occur during a specific time period is known
* each event is independent

**Formula :** 
P(X=k) = lambda k * e-labdda / k!

     - lambda refers to the mean number of events
     - k refers to the number of events
     - e is a constnant equal to approx 2.71828

**Implementation of poisson distribution Using Python :**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average rate of occurrence (lambda)
lambda_ = 4  # average number of visitors per minute

# Possible number of visitors (k values) - from 0 to 15
x = np.arange(0, 16)

# Calculate the PMF for each value of k (number of visitors)
pmf_values = poisson.pmf(x, lambda_)

# Plot the Poisson PMF
plt.bar(x, pmf_values, color='blue', alpha=1)
plt.title('PMF of a Poisson Distribution (lambda=4)')
plt.xlabel('Number of Visitors')
plt.ylabel('Probability')
plt.show()
```

```python
# Calculate the probability of getting exactly 2 visitors
prob_2_visitors = poisson.pmf(2, lambda_)
print(f"Probability of getting exactly 2 visitors: {prob_2_visitors:.4f}")
```

## **PYTHON FOR STATS:**
* SciPy stats
* Statsmodels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df['Sales'].hist()
```

#### checking the empirical rule for this data

```python
mean_sals = df['Sales'].mean()
mean_sals
```

```python
std_sales = df['Sales'].std()
std_sales
```

```python
# 1 std below the mean
lower_limit = mean_sals - (3 * std_sales)
upper_limit = mean_sals + (3 * std_sales)
print(lower_limit,upper_limit)
```

```python
# 1 standard deviation
((df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)).mean()
```

```python
# 2 standard deviation
((df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)).mean()
```

```python
# 3 standard deviation
((df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)).mean()
```

accroding to the python results we can surely say that our data follows the normal distribution
